- Average Dollar volume over last year > 100,000
- Market cap > 10,000,000
- 250 > Price > 5
- Share factor == 1

```python
columns = ["pe", "sharefactor", "sharesbas"]
data = get_data(columns)

data["dollar_vol"] = data["close"] * data["volume"]
# Average dollar volume over last twelve periods
# This is stupid but whatever
dollar_vol_mean = data["dollar_vol"].groupby("ticker").rolling(12).mean().droplevel(0).sort_index()
data["market_cap"] = data["sharesbas"] * data["close"]
data = data[data["sharefactor"] == 1]
data = data[data["close"].between(5, 250)]
data = data[data["market_cap"] > 10000000]
data = data[dollar_vol_mean > 100000]

data = data.dropna()
```
